```python

products = ["t-shirt", "mug", "hat", "book", "keychain"]

customer_orders = set()

inventory = {"t-shirt": 10, "mug" : 5, "hat": 2, "book": 0, "keychain": 15}

```

```python
while True:
    print("Available products:", products)
    product = input("Enter the product you want to order:")
   

    if product in products:
        customer_orders.add(product) 
        print(f"Added {product} to your order.")

    else:
        print("Sorry, we don't have that product.") 
    
    continue_ordering = input("Do you want to order another product? (yes/no): ")
    if continue_ordering.lower() != "yes":
        break
                        
print("Your final  order is:", customer_orders, "Thank you for shopping with us!")

for customer_order in customer_orders:
    if inventory[customer_order] > 0:
        inventory[customer_order] -= 1
        print("{customer_order} is available. Remaining stock: {inventory[customer_order]}")
    else:
        print(f"Sorry, {customer_order} is out of stock.")
```
